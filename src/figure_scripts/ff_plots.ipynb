```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = '/mnt/c/Users/donna/Downloads/Thesis/TSSClassification/results/03-14_11-56_ff_single_coad_full_bed/single_coad_bam--EE87867.hg38.frag.tsv_corrected.bam_fragment_features.csv'
fragment_features = pd.read_csv(file_path)
```

```python
##Removing outlier
Q1 = fragment_features["cov_x"].quantile(0.25)  # First quartile (25%)
Q3 = fragment_features["cov_x"].quantile(0.75)  # Third quartile (75%)
IQR = Q3 - Q1  # Interquartile range

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Keep only values inside the bounds
fragment_features_filtered = fragment_features[(fragment_features["cov_x"] >= lower_bound) & (fragment_features["cov_x"] <= upper_bound)]
```

```python
plt.scatter(fragment_features_filtered['cov_x'], fragment_features_filtered['cov_spread'], s=10)
plt.xlabel("Coverage Depth")
plt.ylabel("Coverage Spread")
plt.grid(True)
plt.show()
```

```python
plt.scatter(fragment_features_filtered['cov_x'], fragment_features_filtered['fslr'], s=5)
plt.xlabel("Coverage Depth")
plt.ylabel("Fragment Short Long Ratio")
plt.grid(True)
plt.show()
```
